```python
import pandas as pd 
import numpy as np 
import seaborn as sn 
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
df = pd.read_csv('laptop_data.csv') 
# check the data 
df.head(5) 
```

```python
# remove the unnamed column 
df = df[['Company', 'TypeName', 'Inches', 'ScreenResolution', 'Cpu', 'Ram', 'Memory', 'Gpu', 'OpSys', 'Weight', 'Price']] 
```

```python
# check for any null values 
df.isnull().sum() 
```

**in this case there are no null values so we can move on** 

```python
df.info() 
```

**This shows that there only the price and inches column is a float, the others are objects**

**we will need to work on some of the columns such as screen resolution , cpu, ram, memory  and weights and make them floats**

```python
'''
in this cell we will treat the weights and Ram columns by replacing the kh in weight with white space and replacing the GB in Ram with white space 
'''
df['Ram'] = df['Ram'].str.replace('GB','')
# converting from string->integer for ram column
df['Ram'] = df['Ram'].astype('int32')

df['Weight'] = df['Weight'].str.replace('kg','')
# converting from string-> float for the weight column
df['Weight'] = df['Weight'].astype('float32')
```

```python
df.info() 
```

**To work on the screen resolution column, check the types of screen first using the value _counts**

```python
df['ScreenResolution'].value_counts()
```

```python
# Then we create another column for touchscreen(1) or not touchscreen (0)
df['TouchScreen'] = df['ScreenResolution'].apply(lambda element:1 if 'Touchscreen' in element else 0 )

#Create a new binary column for IPS facility 
df['IPS'] = df['ScreenResolution'].apply(lambda element:1 if "IPS" in element else 0)
```

```python
df.head() 
```

```python
# extract the resolutions from the screen resolution. 
# we split the tect at the letter 
df['X_Res'] = df['ScreenResolution'].str.extract(r'(\d+)x\d+').astype('int')
df['Y_Res'] = df['ScreenResolution'].str.extract(r'\d+x(\d+)').astype('int')
```

```python
df.head() 
```

```python
df.info() 
```

**Create a new column named `PPI{pixels per inch}`,to combine X-Res and Y_Res. here is the formula of how to calculate `PPI` {pixels per inch}**

$$
    PPI(pixels per inch) = \frac{\sqrt{X_resolution^2+Y_resolution^2}}{inches}
$$

```python
df['PPI'] = (((df['X_Res']**2+df['Y_Res']**2))**0.5/df['Inches']).astype('float')
df.head()
```

```python
# drop the screen resolution, Inches and resolutions 
df.drop(columns=['ScreenResolution','Inches','X_Res','Y_Res'],inplace=True)
df.head()
```

```python
# Check through the CPU column : 
df['Cpu'].value_counts()
```

** From the results above, the following can be observed** 

***1.) Most common processors are made by intel:so we will be clustering their `processors` into different categories*** 

***2.) Any one that dows not fall within the category above will be placed under other interl processor or AMD*** 

***3.) To do this we have to extract the first 3 words***

```python
df['CPU_name'] = df['Cpu'].apply(lambda text:" ".join(text.split()[:3]))
```

```python
'''
As mentioned earlier,if we get any of the intel `i3,i5 or i7` versions
we will return them as it is,but if we get any other processor
we will first check whether is that a variant of the intel? or not
if yes,then we will tag it as "Other Intel Processor" else we will
say it as `AMD Processor`

'''

def processortype(text):
    
    if text=='Intel Core i7' or text=='Intel Core i5' or text=='Intel Core i3':
        return text
    
    else:
        if text.split()[0]=='Intel':
            return 'Other Intel Processor'
        
        else:
            return 'AMD Processor'
        
    
    
df['CPU_name'] = df['CPU_name'].apply(lambda text:processortype(text))
df.head()
```

```python
# Check through the CPU column : 
df['Memory'].value_counts()
```

*Observations:*

**variants observed are HHD, SSD , Flash, Hybrid** 

**we know that ITB ==> 1000GB**

**We will create four columns for each variants and keep the units in GB** 


```python
# change the 1.0TB to 1TB 
df['Memory'] = df['Memory'].astype(str).replace('\.0','', regex=True) 

# replace the TB word with "000"
df['Memory'] = df['Memory'].str.replace('TB', '000') 

# replace the GB word with " "
df['Memory'] = df['Memory'].str.replace('GB','')

#split the word accross the "+" character

newdf = df['Memory'].str.split("+",n = 1,expand = True)

newdf
```

```python
# Strip up all the white spaces.
df['first'] = newdf[0]
df['first'] = df['first'].str.strip()
df.head()
```

```python
def applychanges(value):
    df['Layer1'+value] = df['first'].apply(lambda x:1 if value in x else 0)


listtoapply = ['HDD','SSD','Hybrid','FlashStorage']    
for value in listtoapply:
    applychanges(value)
        
df.head()
```

```python
# remove all the characters just keep the numbers
df['first'] = df['first'].str.replace(r'\D','')
df['first'].value_counts()
```

```python
# take the second column 
df['Second'] = newdf[1]
df.head()
```

```python
def applychanges1(value):
    df['Layer2'+value] = df['Second'].apply(lambda x:1 if value in x else 0)
    
    
listtoapply1 = ['HDD','SSD','Hybrid','FlashStorage']
df['Second'] = df['Second'].fillna("0")
for value in listtoapply1:
    applychanges1(value)
```

```python
# remove all the characters just keep the numbers

df['Second'] = df['Second'].str.replace(r'\D','')
df['Second'].value_counts()
```

```python
# change both to integers 
df['first'] = df['first'].astype('int')
df['Second'] = df['Second'].astype('int')
df.head()
```

```python
# multiplying the elements and storing the result in subsequent columns
df["HDD"]=(df["first"]*df["Layer1HDD"]+df["Second"]*df["Layer2HDD"])
df["SSD"]=(df["first"]*df["Layer1SSD"]+df["Second"]*df["Layer2SSD"])
df["Hybrid"]=(df["first"]*df["Layer1Hybrid"]+df["Second"]*df["Layer2Hybrid"])
df["Flash_Storage"]=(df["first"]*df["Layer1FlashStorage"]+df["Second"]*df["Layer2FlashStorage"])
```
